# Real Estate House Price Prediction

#### Building a Linear Regression model

### Import the necessary libraries

```python
#import seaborn
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Read the csv data into a pandas dataframe and display the first 5 samples

```python
df = pd.read_csv('Real_estate.csv')
df.head()
```

### Show more information about the dataset

```python
df.info()
```

### Find how many samples are there and how many columns are there in the dataset

```python
df.shape
```

### What are the features available in the dataset?

```python
df.info()
```

### Check if any features have missing data

```python
df.isnull().sum()
```

### Group all the features as independent features in X

```python
X = df[['No', 'X1 transaction date', 'X2 house age', 'X3 distance to the nearest MRT station', 'X4 number of convenience stores', 'X5 latitude', 'X6 longitude']]
X
```

### Group feature(s) as dependent features in y

```python
y = df.drop(columns = ['No', 'X1 transaction date', 'X2 house age', 'X3 distance to the nearest MRT station', 'X4 number of convenience stores', 'X5 latitude', 'X6 longitude'])
y
```

### Split the dataset into train and test data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
```

### Choose the model (Linear Regression)

```python
from sklearn.linear_model import LinearRegression
```

### Create an Estimator object

```python
reg = LinearRegression()
```

### Train the model

```python
model = reg.fit(X_train, y_train)
model
```

### Apply the model

```python
model = reg.predict(X_test)
model
```

### Display the coefficients

```python
reg.coef_
```

### Find how well the trained model did with testing data

```python
reg.score(X_test,y_test)
```
